```python
import requests
import itertools
import os
import csv

# get list of symbols
with open(r'C:\Users\lalit\Desktop\w266\transcripts-symbol.csv', 'r') as f:
    symbols = f.read().split('\n')
    
print(symbols[:20])
print('number of symbols is', len(symbols))

exp_symbols = []
for symbol in symbols[:100]:
    # split the symbols using , and replace the " with blank
    new_symbol = symbol.replace('"', '').split(',')
    exp_symbols.append(new_symbol)

exp_symbols_flat = list(itertools.chain.from_iterable(exp_symbols))
final_symbols = list(set(exp_symbols_flat))

print(final_symbols[:20])
print('number of exp_symbols is', len(final_symbols))

# get the ids for each symbol
a1 = 'https://finnhub.io/api/v1/stock/transcripts/list?symbol='
a3 = '&token=c0u0c1n48v6qqphtp4c0'

ids = []
transcripts_ids = []
transcripts_times = []

for symbol in final_symbols:
    # print(a1+symbol+a3)
    ids.append(requests.get(a1+symbol+a3).json())

    for each_id in ids:
        try:
            for entry in each_id['transcripts']:
                # print('able to get access for', each_id)
                transcripts_ids.append(entry['id'])
                transcripts_times.append(entry['time'])
        except:
            # print('unable to find keyword "transcripts" in id', each_id)
            continue
        
# print(transcripts_ids, transcripts_times)
print('number of ids is', len(transcripts_ids))
```

```python
# get the transcripts for each id
b1 = 'https://finnhub.io/api/v1/stock/transcripts?id='
b3 = '&token=c0u0c1n48v6qqphtp4c0'

transcripts = []
final = {}

for index, each_id in enumerate(transcripts_ids[:10000]):
    try:
        x = requests.get(b1+each_id+b3).json()['transcript']
        transcripts.append(x)
    
        s = []
        for each_transcript in x:
            s.append(each_transcript['speech'])

        merged = list(itertools.chain.from_iterable(s))
        d = {(each_id, transcripts_times[index]): ''.join(merged)}
        final.update(d)

    except:
        continue

print(len(final))
```
